```python
import pandas as pd
import numpy as np
import datetime as dt
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
```

## Task 1 Compound Return
you need the 49_Industry_Portfolios_Daily.csv1 in the raw data folder
provided to you. In this data file, you will find both the value-weighted and the equal-weighted
returns for 49 industry portfolios. You only need the value-weighted returns. You do NOT need the
equal-weighted returns throughout this data task.
Based on value weighted returns, how much money will you have at the end of June 29th, 2007 if you
put 1 dollar in the banking industry portfolio at the beginning of January 3rd, 1994?

```python
df1 = pd.read_csv('49_Industry_Portfolios_Daily.csv', skiprows = 9, nrows = 25105)
df1.rename(columns={'Unnamed: 0':'Date'}, inplace = True)

df1_slice = df1[df1['Date'] <= 20070629]
df1_slice = df1_slice[df1_slice['Date'] >= 19940103]
df1_slice['bank_log_return'] = df1_slice['Banks'].apply(lambda x: np.log(1+x/100))
bank_compound_return = np.exp(df1_slice['bank_log_return'].sum())
print("%.4f" % bank_compound_return)
```

## Task 2: Two-Day Return
Please calculate the 2-day return for each of the industry portfolios and the market portfolio at
each FOMC meeting event. Here, we define the 2-day return as the compound return from the
beginning of the FOMC meeting day T to the end of day T + 1. For example, if your put 1 dollar
in the banking industry portfolio at the beginning of day T, and get X dollars at the end of day
T + 1, we say the 2-day return for banking industry is X.
What is the average 2-day returns for the agriculture industry portfolio around all the scheduled
FOMC meetings from January 3rd, 1994 to June 29th, 2007?

```python
df2 = pd.read_csv('F-F_Research_Data_Factors_daily.csv', skiprows = 4, nrows = 25105)
df2.rename(columns={'Unnamed: 0':'Date'}, inplace = True)
df2 = df2[['Date','Mkt-RF']]
df1 = df1.merge(df2, how = 'left', on = 'Date')
df3 = pd.read_excel('daily_FF_surprises_July_2008.xls', sheet_name = 'Event days')
df3.rename(columns={'Unnamed: 0':'Date'}, inplace = True)
df3 = df3[['Date','Scheduled FOMC meeting']]
df1['Date'] = df1.Date.astype(str)
df1['Date'] = pd.to_datetime(df1['Date'])
df1 = df1.merge(df3, how = 'left', on = 'Date')
df1['Scheduled FOMC meeting'].value_counts()

df4 = pd.read_excel('replication_dataset_gw.xlsx', sheet_name = 'Sheet2')
df4.rename(columns={'DATE':'Date'}, inplace = True)
df4 = df4[['Date','tight_suprise']]
df1 = df1.merge(df4, how = 'left', on = 'Date')
df1['tight_suprise'].value_counts()

start_date = dt.datetime(1994,1,3)
end_date = dt.datetime(2007,6,29)
df = df1[(df1['Date'] >= start_date) & (df1['Date'] <= end_date)]

df_2day = pd.DataFrame(columns = df.columns)
df_2day['Date'] = df['Date']
list = df1.columns
for index, row in df.iterrows():
    if row['Scheduled FOMC meeting'] == 1:
        for item in list:
            if (item != 'Date') & (item != 'Scheduled FOMC meeting') & (item != 'tight_suprise'):
               df_2day.loc[index, item] = (1+df.loc[index][item]/100)*(1+df.loc[index+1][item]/100)
df_2day.dropna(inplace =True, how = 'all', axis = 1)               
df_2day.dropna(inplace = True, how = 'any', axis = 0)
avg_agric_2day_return = df_2day['Agric'].mean()
print("%.4f" % avg_agric_2day_return)
df_2day = df_2day.merge(df4, how = 'left', on = 'Date')
```

## Task 3: Plot the Coefficients
For each industry and the market, regress the 2-day returns on “tight surprise” variable for scheduled
meetings only (i.e., use OLS regressions; the y-variable should be the 2-day return; the x-variable
should be the tight surprise). You may want to use a loop to do these regressions (i.e., 50 regressions
in total; one regression for each industry; plus a regression for the market). Then plot the coefficient
of each industry and the coefficient of the market in a bar graph. Please sort the coefficients in
increasing order and label each industry.

```python
names = df_2day.columns.values.tolist()
del names[0],names[50]
values= []
reg = linear_model.LinearRegression()
x = df_2day['tight_suprise'].to_numpy().reshape(-1,1)
for item in names:
    y = df_2day[item].to_numpy().reshape(-1,1)
    reg.fit(x,y)
    values.append(reg.coef_[0][0])

plt_df = pd.DataFrame(columns = ['name','value'])
plt_df['name'] = names
plt_df['value'] = values
plt_df.sort_values('value', inplace = True)
f, ax= plt.subplots(figsize = (14, 10))
sns.set_theme(style="whitegrid")
clrs = ['red' if (name == 'Banks') or (name == 'Mkt-RF') else 'blue' for name in plt_df['name'] ]
sns.barplot(x = 'name', y = 'value', data = plt_df, palette = clrs)
ax.tick_params(axis='x',labelsize=12)
ax.set_xticklabels(ax.get_xticklabels(), rotation=270)
plt.show()
```
